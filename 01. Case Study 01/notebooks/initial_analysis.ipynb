```python
# Criar um dataframe para cada planilha.
file_path_q1 = '/kaggle/input/clean-divvy-data/data_q1_editado.csv'
file_path_q2 = '/kaggle/input/clean-divvy-data/data_q2_editado.csv'
file_path_q3 = '/kaggle/input/clean-divvy-data/data_q3_editado.csv'
file_path_q4 = '/kaggle/input/clean-divvy-data/data_q4_editado.csv'

# Carregar os dados.
data_q1 = pd.read_csv(file_path_q1)
data_q2 = pd.read_csv(file_path_q2)
data_q3 = pd.read_csv(file_path_q3)
data_q4 = pd.read_csv(file_path_q4)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Converter colunas de data para o tipo datetime se ainda não estiverem
data_q1['start_time'] = pd.to_datetime(data_q1['start_time'])
data_q2['start_time'] = pd.to_datetime(data_q2['start_time'])
data_q3['start_time'] = pd.to_datetime(data_q3['start_time'])
data_q4['start_time'] = pd.to_datetime(data_q4['start_time'])

# Criar coluna 'hour' para representar a hora do dia
data_q1['hour'] = data_q1['start_time'].dt.hour
data_q2['hour'] = data_q2['start_time'].dt.hour
data_q3['hour'] = data_q3['start_time'].dt.hour
data_q4['hour'] = data_q4['start_time'].dt.hour

# Configurações visuais
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Plotar padrões de uso ao longo do dia
plt.figure(figsize=(14, 16))

plt.subplot(4, 1, 1)
sns.countplot(x='hour', data=data_q1, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q1')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 2)
sns.countplot(x='hour', data=data_q2, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q2')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 3)
sns.countplot(x='hour', data=data_q3, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q3')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 4)
sns.countplot(x='hour', data=data_q4, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q4')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.tight_layout()
plt.savefig('padroes_de_uso_aprimorado.png')
plt.show()

# Seu código para preparar e plotar os padrões de uso ao longo do dia...

plt.savefig('padroes_de_uso.png')
plt.show()
```

```python
# Concatenar os dataframes
all_data = pd.concat([data_q1, data_q2, data_q3, data_q4])
all_data.describe()
```

```python
# Informações básicas sobre o novo DataFrame
print("1. Nomes das Variáveis (Colunas):")
print(all_data.columns.tolist())

print("\n2. Tipos de Dados:")
print(all_data.dtypes)

print("\n3. Estatísticas Descritivas:")
print(all_data.describe())

print("\n6. Tamanho do DataFrame:")
print(all_data.shape)
```

```python
# Filtrar coluna 'birthyear' para valores não nulos e maiores que 1940
filtered_data = all_data[(all_data['birthyear'].notnull()) & (all_data['birthyear'] >= 1940)]

# Configurar a paleta de cores
sns.set_theme(style="whitegrid", palette="viridis")

# Criar histograma
plt.figure(figsize=(12, 6))
sns.histplot(filtered_data['birthyear'], bins=30, kde=False)
plt.title('Distribuição Etária dos Usuários (Nascidos a partir de 1940)')
plt.xlabel('Ano de Nascimento')
plt.ylabel('Número de Usuários')
plt.savefig('birthyear_x_user.png')
plt.show()

# Calcular estatísticas descritivas
mean_age = filtered_data['birthyear'].mean()
median_age = filtered_data['birthyear'].median()

print(f'Média de Idade: {mean_age:.2f}')
print(f'Mediana de Idade: {median_age:.2f}')
```

```python
# Agrupar por tipo de usuário e gênero e calcular estatísticas descritivas da idade
age_stats = all_data.groupby(['usertype', 'gender'])['birthyear'].agg(['mean', 'median']).reset_index()

# Renomear colunas para clareza
age_stats.columns = ['usertype', 'gender', 'mean_age', 'median_age']

# Exibir estatísticas descritivas
print(age_stats)
```

```python
# Leitura dos DataFrames
data_q1 = pd.read_csv(file_path_q1)
data_q2 = pd.read_csv(file_path_q2)
data_q3 = pd.read_csv(file_path_q3)
data_q4 = pd.read_csv(file_path_q4)

# Concatenação vertical dos DataFrames
all_data = pd.concat([data_q1, data_q2, data_q3, data_q4], ignore_index=True)

# Verificando o novo DataFrame
print(all_data.head())

# Salvar o DataFrame resultante em um novo arquivo, se necessário
all_data.to_csv('all_data.csv', index=False)
```

```python
# Configuração do estilo
sns.set_theme(style="whitegrid", palette="viridis")

# Contagem de viagens para cada combinação de gênero e tipo de usuário
gender_user_count = all_data.groupby(['gender', 'usertype']).size().reset_index(name='count')

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='gender', y='count', hue='usertype', data=gender_user_count)

# Adicionar rótulos às barras
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center', xytext = (0, 10),
                     textcoords = 'offset points')

plt.title('Relação entre Gênero e Tipo de Usuário')
plt.xlabel('Gênero')
plt.ylabel('Número de Viagens')

# Salvar as imagens
plt.savefig('genero_x_usuario')

# Exibir o gráfico
plt.show()

```

```python
# Converter 'start_time' para tipo datetime
all_data['start_time'] = pd.to_datetime(all_data['start_time'])

# Extrair o mês de 'start_time'
all_data['month'] = all_data['start_time'].dt.month

# Criar gráfico de barras empilhadas para análise sazonal
plt.figure(figsize=(14, 8))
sns.countplot(x='month', hue='usertype', data=all_data, palette='viridis')
plt.title('Análise Sazonal por Tipo de Usuário')
plt.xlabel('Mês')
plt.ylabel('Número de Viagens')
plt.show()
```

```python
# Filtrar dados para anos de nascimento após 1950
filtered_data_age = all_data[all_data['birthyear'] >= 1950]

# Criar histograma para a distribuição de idade por tipo de usuário
plt.figure(figsize=(14, 8))
sns.histplot(data=filtered_data_age, x='birthyear', hue='usertype', multiple='stack', bins=30, palette='viridis')
plt.title('Distribuição de Idade por Tipo de Usuário (Nascidos após 1950)')
plt.xlabel('Ano de Nascimento')
plt.ylabel('Número de Usuários')
plt.show()
```

```python
# Calcular contagem de viagens para cada destino por tipo de usuário
destinations_year = all_data.groupby(['to_station_name', 'usertype']).size().reset_index(name='count')

# Filtrar os 10 destinos mais populares para cada tipo de usuário
top_destinations_subscriber_year = destinations_year[destinations_year['usertype'] == 'Subscriber'].nlargest(10, 'count')
top_destinations_customer_year = destinations_year[destinations_year['usertype'] == 'Customer'].nlargest(10, 'count')

# Criar gráficos de barras separados para cada tipo de usuário
plt.figure(figsize=(16, 10))

plt.subplot(2, 1, 1)
sns.barplot(x='count', y='to_station_name', data=top_destinations_subscriber_year, palette='viridis', order=top_destinations_subscriber_year['to_station_name'])
plt.title('Top 10 Destinos para Usuários Assinantes')

plt.subplot(2, 1, 2)
sns.barplot(x='count', y='to_station_name', data=top_destinations_customer_year, palette='viridis', order=top_destinations_customer_year['to_station_name'])
plt.title('Top 10 Destinos para Usuários Comuns')

# Ajustes adicionais para legibilidade
plt.xlabel('Número de Viagens')
plt.ylabel('Estação de Destino')
plt.tight_layout()

# Salvar as imagens
plt.savefig('top_destinos_por_user_year.png')
plt.show()
```
